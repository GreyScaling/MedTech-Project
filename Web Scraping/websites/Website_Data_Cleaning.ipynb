```python
for condition in data['Condition']:
    
    if "healthy" in condition.lower():
        data['Category'] = "Healthy"
        
    elif "covid" in condition.lower():
        data['Category'] = "Covid"
        
    elif "brain" in condition.lower():
        data['Category'] = "Brain Diseases"
        
    elif ('stroke' or 'heart') in condition.lower():
        data['Category'] = "Heart Diseases"
        
    elif ('stress' or 'anxiety' or 'depression') in condition.lower():
        data['Category'] = "Mental Health"
    
    elif ('obesity' or 'diabete' or 'mellitus') in condition.lower():
        data['Category'] = "Obesity"
    
    elif ('shoulder' or 'back' or 'neck') in condition.lower():
        data['Category'] = "Body Aches"
        
    elif ('respiratory' or 'lung' or 'breathing') in condition.lower():
        data['Category'] = 'Respiratory Diseases'
        
    else:
        data['Category'] = "Others"
```
